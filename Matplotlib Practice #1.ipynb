## -- Visualisation using Matplotlib --

```python
#Import the required libraries

import matplotlib.pyplot as plt
import pandas as pd, numpy as np
```

### - Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

#Resize the graph
plt.figure(figsize=(5,3), dpi=300)

##Line 1

#Keyword argument notation
#plt.plot(x, y, label='2x', color='red', linewidth=2, markersize=2)

#Short hand notation for line restyle
#fmt = '[color][marker][line]'
plt.plot(x, y, 'b^--', label='2x')

##Line 2

#Create a second line
x2 = np.arange(0,4.5,0.5)

#Plot part of the graph as a line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

#Plot part of the graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

#Add a title and labels
plt.title('Our first graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

#Add X and Y tick marks
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

#Add a legend
plt.legend()

#Saving the graph
plt.savefig('mygraph.png', dpi=500)

#Show plot
plt.show()
```

### - Bar Charts

```python
#Creating lists of values
labels = ['A','B','C']
values = [1,4,2]

#Plot a bar chart
bars = plt.bar(labels, values)

#Creating a for loop for setting hatch
patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('o')
# bars[2].set_hatch('*')

#Resize chart
plt.figure(figsize= (6,4))

#Show plot
plt.show()
```
